# **2. Model Training: Building a churn prediction model**

Now we are ready to enter in the training stage of the machine learning models. The common way to procedure is starting with baseline models (i.e. SVM, Decision Trees, Naive Bayes, etc....) and later, try to improve it adjusting hyperparameters of the models or creating more complex models architectures as ensembles. In this case we will only improve the model by adjusting its hyperparameters. 

There is a very important step when building a model, and it is to choose a right model threshold (usually known as model cutoff value). This threshold modifies the output class probability density function in order to balance the model and fit our requirements. for example, imagine a case in which you have to design a model to detect sick people in order to speed up their medical treatment. Is it better either to classify healthy people as sick people or sick people as healthy ones? Normally, it is better to classify healthy people as sick ones, because they will be removed after the medical tests, than sick people as healthy ones, because they won't be treated and maybe will die ... Therefore, the model cutoff plays a really important role when designing a model. Some libraries already has the cutoff as an input parameter, but normally it is not and by default is assigned to 0.5. In this case, what you have to do is to train your model, predict (using probabilities) and modify the prediction value according to your threshold.

[**EX 10**]

The function *train_model* is a function that trains the model and plots the most interesting, and important, metrics that will help you to rightly define the model hyperparameters. **Your task here is to fill those lines that begins with TODO**

**Solution**

```python
def train_model_and_performance_eval(logit,train_x,test_x,train_y,test_y, cols, cutoff = 0.5, cf = 'coefficients'):
    # TODO: fit the model (train)
    logit.fit(train_x, train_y)
    
    # TODO: calculate its predictions (0, 1) and their probability 
    predictions = logit.predict(test_x)
    probabilities = logit.predict_proba(test_x)

    
    # TODO: modify the prediction value according to the cutoff value
    predictions = np.where(probabilities[:,1] >= cutoff, 1, 0)
    
    
    # Calculate the coefficients dataframe depending on if they are real coefficients or features
    if cf == "coefficients" :
        coefficients  = pd.DataFrame(logit.coef_.ravel())
    elif cf == "features" :
        coefficients  = pd.DataFrame(logit.feature_importances_)
        
    
    # Set the coefficients to be shown
    column_df     = pd.DataFrame(cols)
    coef_sumry    = (pd.merge(coefficients,column_df,left_index= True,
                              right_index= True, how = "left"))
    coef_sumry.columns = ["coefficients","features"]
    coef_sumry    = coef_sumry.sort_values(by = "coefficients",ascending = False)
    
    # TODO: print the accuracy and classification reports (see accuracy_score and classification_report functions)
    acc_score = accuracy_score(test_y, predictions)
    print(f'Accuracy: {acc_score}')
    class_report = classification_report(test_y, predictions)
    print(f'Classification report: \n {class_report}')
    
    # TODO: Plot confusion matrix (see plot_confusion_matrix function)  
    plot_confusion_matrix(logit, test_x, test_y)

    # TODO: Plot the probability density chart 
    
    churn_prob = probabilities[:,1]

    fig,axes=plt.subplots(1,2, figsize=(10, 5))

    sns.distplot(churn_prob[test_y == 1],color="lightblue", ax=axes[0]) 
    sns.distplot(churn_prob[test_y == 0],color="gold", ax=axes[1])  
    
    
    fig.legend(labels=['Churn','No Churn'])
    fig.suptitle('Probability density chart')
    plt.show()
    
    fig = plt.figure(figsize=(15,5))
    sns.distplot(churn_prob[test_y == 0], color="gold")
    sns.distplot(churn_prob[test_y == 1], color="lightblue")
    fig.legend(labels=['No Churn','Churn'])
    fig.suptitle('Probability density chart')
    plt.axvline(x=cutoff, c='orange')
    plt.show()
    # Plot feature importance bar
    # Prepare Data
    coef_sumry.reset_index(inplace=True)
    coef_sumry['colors'] = ['red' if x < 0 else 'green' for x in coef_sumry['coefficients']]
    coef_sumry

    # Draw plot
    plt.figure(figsize=(10,5), dpi= 80)
    plt.hlines(y=coef_sumry.index, xmin=0, xmax=coef_sumry.coefficients, color=coef_sumry.colors, alpha=0.4, linewidth=5)
        
    # Decorations
    plt.gca().set(ylabel='$Features$', xlabel='$Coefficients$')
    plt.yticks(coef_sumry.index, coef_sumry.features, fontsize=12)
    plt.title('Feature Importance', fontdict={'size':20})
    plt.grid(linestyle='--', alpha=0.5)
    plt.show()
```

# **3. Lead Generation Pipeline**

For our case, we have been required to build up a Lead Generation Pipeline for potential churners that have hired a streaming service with our Telco company. In order to be able to build up the pipeline, Business provided the business rules that define how we must proceed:

**Leads classification**

Leads will be classified into three groups: Tier 1, Tier 2 and Tier 3, where Tier 1 belongs to best leads and Tier 3 for worst ones.

**Tiers assignation**

- **Tier 1**

  - Has Streaming TV and Movies 
  - Has any internet service
  - Tenure >= 12
  - Monthly charges >= 75


- **Tier 2**

  - Has Streaming TV or Movies 
  - Has any internet service
  - Tenure >= 12
  - Monthly charges >= 50
  

- **Tier 3**

  - Has Streaming TV or Movies 
  - The rest

[**EX 13**]

Fill up the function *lead_generation_pipeline* which performs the Lead Generation Pipeline. Follow the indications above in order to assign the Tier values correctly. The output must be the input with a new column (Tier) that contains the assigned tier value.

**Solution**

```python
def lead_generation_pipeline(data: pd.DataFrame):
    # Filter to only those that are churners
    data = data[data.PossibleChurner == 1]

    tier1 = data[((data.StreamingTV == 'Yes') & (data.StreamingMovies == 'Yes')) & 
                (data.InternetService != 'No') & (data.tenure >= 12) & (data.MonthlyCharges >= 75)]
    
    tier2 = data[((data.StreamingTV == 'Yes') | (data.StreamingMovies == 'Yes')) & 
                (data.InternetService != 'No') & (data.tenure >= 12) &
                (data.MonthlyCharges >= 50)]
    
    tier3 = data[((data.StreamingTV == 'Yes') | (data.StreamingMovies == 'Yes'))]
   
    # TODO: build up the output dataframe by following the rules above
    output = pd.concat([tier1, tier2, tier3]).drop_duplicates()
    
    
    output.loc[output['customerID'].isin(tier3.customerID), "Tier"] = "Tier 3"
    output.loc[output['customerID'].isin(tier2.customerID), "Tier"] = "Tier 2"
    output.loc[output['customerID'].isin(tier1.customerID), "Tier"] = "Tier 1"
    
    return output
```

```python
output = lead_generation_pipeline(data = original_test)
```
